```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
prefix = 'checkpoints/r18_fisher_'
params = [11680678, 8952944, 3500000, 780000]

fig, ax = plt.subplots(1,1,figsize=(10,5))

for line in ['ltick', 'scratch']:
    x = params
    y = []
    for p in params:
        state = torch.load(prefix + line + '_' + str(p) + '_params.t7', map_location='cpu')
        print(state.keys())
        min_err = min(state['error_history'])
        print(len(state['error_history']))
        print(min_err)
        y.append(min_err)
    
    ax.plot(x,y,marker='o')
    
ax.legend(['ltick', 'scratch'])
        
plt.show()
```
